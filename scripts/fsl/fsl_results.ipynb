```python
# Daniel Bandala @ nov-2022
# general libraries
import numpy as np
import matplotlib.pyplot as plt
# diffussion image processing
from dipy.io.image import load_nifti
```

```python
# set plot parameters
plt.rcParams.update({'font.size': 20})
```

```python
# folders dataset
main_dir = '/content/drive/MyDrive/dti-transformer/dti_data/HCP'
```

```python
i = 2
# read difussion data
data, affine = load_nifti(main_dir+'/case_0'+str(i)+'/input/DWI.nii.gz')
# read difussion data
fa1, _ = load_nifti(main_dir+'/case_0'+str(i)+'/dti/DTI_FA.nii.gz')
md1, _ = load_nifti(main_dir+'/case_0'+str(i)+'/dti/DTI_MD.nii.gz')
mo1, _ = load_nifti(main_dir+'/case_0'+str(i)+'/dti/DTI_MO.nii.gz')
# read difussion data
fa2, _ = load_nifti(main_dir+'/case_0'+str(i+1)+'/dti/DTI_FA.nii.gz')
md2, _ = load_nifti(main_dir+'/case_0'+str(i+1)+'/dti/DTI_MD.nii.gz')
mo2, _ = load_nifti(main_dir+'/case_0'+str(i+1)+'/dti/DTI_MO.nii.gz')
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2, 3, 1)
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.imshow(fa1[:, :, 48], cmap='gray', origin='lower')
plt.ylabel("Caso 1")
plt.title("Anisotropía fraccional")
plt.subplot(2, 3, 2).set_axis_off()
plt.imshow(md1[:, :, 48], cmap='gray', origin='lower')
plt.title("Difusividad media")
plt.subplot(2, 3, 3).set_axis_off()
plt.imshow(mo1[:, :, 48], cmap='gray', origin='lower')
plt.title("Moda de anisotropia")
plt.subplot(2, 3, 4)
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.imshow(fa2[:, :, 48], cmap='gray', origin='lower')
plt.ylabel("Caso 2")
plt.subplot(2, 3, 5).set_axis_off()
plt.imshow(md2[:, :, 48], cmap='gray', origin='lower')
plt.subplot(2, 3, 6).set_axis_off()
plt.imshow(mo2[:, :, 48], cmap='gray', origin='lower')
plt.savefig("fsl_result_"+str(i)+".png")
```
