```python
# Daniel Bandala @ nov-2022
# general libraries
import numpy as np
import matplotlib.pyplot as plt
# diffussion image processing
from dipy.io.image import load_nifti
```

```python
# set plot parameters
plt.rcParams.update({'font.size': 22})
```

```python
# folders dataset
main_dir = '/content/drive/MyDrive/dti-transformer/dti_data/HCP'
```

```python
i = 16
# read difussion data
fa1, _ = load_nifti(main_dir+'/case_'+str(i)+'/output/DTI_FA.nii.gz')
md1, _ = load_nifti(main_dir+'/case_'+str(i)+'/dti/DTI_MD.nii.gz')
#mo1, _ = load_nifti(main_dir+'/case_'+str(i)+'/dti/DTI_MO.nii.gz')
rgb1, _ = load_nifti(main_dir+'/case_'+str(i)+'/output/DTI_RGB.nii.gz')
# read difussion data
fa2, _ = load_nifti(main_dir+'/case_'+str(i+2)+'/output/DTI_FA.nii.gz')
md2, _ = load_nifti(main_dir+'/case_'+str(i+2)+'/dti/DTI_MD.nii.gz')
#mo2, _ = load_nifti(main_dir+'/case_'+str(i+1)+'/dti/DTI_MO.nii.gz')
rgb2, _ = load_nifti(main_dir+'/case_'+str(i+2)+'/output/DTI_RGB.nii.gz')
fa1.shape
```

```python
s = 48
plt.figure(figsize=(18,12))
plt.subplot(2, 3, 1)
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.imshow(np.rot90(fa1[:, :, s], k=1), cmap='gray', origin='lower')
plt.title("Anisotropía fraccional")
plt.subplot(2, 3, 2).set_axis_off()
plt.imshow(np.rot90(md1[:, :, s], k=1), cmap='gray', origin='lower')
plt.title("Difusividad media")
plt.subplot(2, 3, 3).set_axis_off()
plt.imshow(np.rot90(rgb1[:, :, s], k=1), cmap='gray', origin='lower')
plt.title("Orientación de difusión")
plt.subplot(2, 3, 4)
plt.gca().set_yticklabels([])
plt.gca().set_xticklabels([])
plt.gca().set_xticks([])
plt.gca().set_yticks([])
plt.imshow(np.rot90(fa2[:, :, s], k=1), cmap='gray', origin='lower')
plt.subplot(2, 3, 5).set_axis_off()
plt.imshow(np.rot90(md2[:, :, s], k=1), cmap='gray', origin='lower')
plt.subplot(2, 3, 6).set_axis_off()
plt.imshow(np.rot90(rgb2[:, :, s], k=1), cmap='gray', origin='lower')
plt.tight_layout()
plt.savefig("fsl_result_"+str(i)+".png")
```
